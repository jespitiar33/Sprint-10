## Parte 1. Priorizar hipótesis

El archivo hypotheses_us.csv contiene nueve hipótesis sobre cómo aumentar los ingresos de una tienda online con Reach, Impact, Confidence y Effort especificados para cada una.

El ejercicio consiste en:

Aplicar el framework ICE para priorizar hipótesis. Ordenarlas en orden descendente de prioridad.
Aplicar el framework RICE para priorizar hipótesis. Ordenarlas en orden descendente de prioridad.
Mostrar cómo cambia la priorización de hipótesis cuando utilizas RICE en lugar de ICE. Proporcionar una explicación de los cambios.

```python
import pandas as pd

# Almacena los datos de visitas, pedidos y gastos en variables.
hypotheses = pd.read_csv('/datasets/hypotheses_us.csv', sep=';')
print(hypotheses)
print()
print(hypotheses.info())
```

```python
hypotheses.columns = ['hypothesis', 'reach', 'impact', 'confidence', 'effort']
print(hypotheses.head(3))
```

```python
# Cambiar los datos de las columnas 'reach', 'impact', 'confidence' y 'effort' de string a int

hypotheses['reach'] = hypotheses['reach'].astype(int)
hypotheses['impact'] = hypotheses['impact'].astype(int)
hypotheses['confidence'] = hypotheses['confidence'].astype(int)
hypotheses['effort'] = hypotheses['effort'].astype(int)
```

<div class="alert alert-block alert-success"> <b>Celda [1-3]</b> <a class="tocSkip"></a> La carga del archivo CSV y la transformación de los nombres de columnas son correctas. Esto facilita el acceso a los datos y evita errores en operaciones posteriores.</div> <div class="alert alert-block alert-warning"> <b>Celda [3]</b> <a class="tocSkip"></a> Convertir los valores de las columnas 'reach', 'impact', 'confidence' y 'effort' a `int` es innecesario, ya que ya están en formato `int64`. Puedes eliminar estas líneas para mejorar la eficiencia.</div>

```python
# Aplicar el framework ICE para priorizar hipótesis. Ordenarlas en orden descendente de prioridad.

hypotheses['ICE'] = (hypotheses['impact']*hypotheses['confidence'])/hypotheses['effort']
hypotheses = hypotheses.sort_values(by='ICE', ascending=False) 
print(hypotheses.head(10))
```

```python
# Aplicar el framework RICE para priorizar hipótesis. Ordenarlas en orden descendente de prioridad.

hypotheses['RICE'] = (hypotheses['reach']*hypotheses['impact']*hypotheses['confidence'])/hypotheses['effort']
hypotheses = hypotheses.sort_values(by='RICE', ascending=False) 
print(hypotheses[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False))
```

<div class="alert alert-block alert-success"> <b>Celda [4-5]</b> <a class="tocSkip"></a> La implementación de los frameworks **ICE y RICE** es correcta. Se usan fórmulas adecuadas y la ordenación es coherente.</div> <div class="alert alert-block alert-warning"> <b>Celda [5]</b> <a class="tocSkip"></a> El cálculo de **RICE** puede optimizarse evitando la división por cero si hay valores de `effort` en 0 (aunque aquí no ocurra, es buena práctica manejar excepciones):

    hypotheses['RICE'] = np.where(
        hypotheses['effort'] == 0, 
        0, 
        (hypotheses['reach'] * hypotheses['impact'] * hypotheses['confidence']) / hypotheses['effort']
    )
</div>

```python
# Mostrar cómo cambia la priorización de hipótesis cuando utilizas RICE en lugar de ICE.

hypotheses_ICE_vs_RICE = hypotheses[['hypothesis', 'RICE', 'ICE']].sort_values(by='RICE', ascending=False)
print(hypotheses_ICE_vs_RICE)
```

```python
# Gráfica 'hypotheses_ICE_vs_RICE'

import matplotlib.pyplot as plt

hypotheses_ICE_vs_RICE.plot(kind='bar', x='hypothesis', y=['ICE', 'RICE'], stacked=True)
plt.show()
```

## Parte 2. Análisis del test A/B

```python
orders = pd.read_csv('//datasets/orders_us.csv')
print(orders.head(5))
print(orders.info())
print()
visits = pd.read_csv('/datasets/visits_us.csv')
print(visits.head(5))
print(visits.info())
```

<div class="alert alert-block alert-success"> <b>Celda [8]</b> <a class="tocSkip"></a> La carga de los datasets `orders` y `visits` se realiza correctamente. También verificaste los tipos de datos con `.info()`, lo cual es una buena práctica.</div> <div class="alert alert-block alert-warning"> <b>Celda [9]</b> <a class="tocSkip"></a> No es necesario convertir `date` a `str`, ya que después lo transformas a `datetime`. Puedes eliminar esta conversión intermedia.</div>

```python
# Preprocesar los datos.

# Cambiamos el tipo de dato de la columna 'date' a string
orders['date'] = orders['date'].astype(str)
visits['date'] = visits['date'].astype(str)
```

```python
import datetime as dt
import numpy as np

orders['date'] = orders['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
visits['date'] = visits['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
```

```python
datesGroups = orders[['date','group']].drop_duplicates()
```

```python
datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])].agg({
  'date' : 'max',
  'group' : 'max',
  'transactionId' : pd.Series.nunique,
  'visitorId' : pd.Series.nunique,
  'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])

ordersAggregated = datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'transactionId' : pd.Series.nunique, 'visitorId' : pd.Series.nunique, 'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])

visitsAggregated = datesGroups.apply(lambda x: visits[np.logical_and(visits['date'] <= x['date'], visits['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'visits' : 'sum'}), axis=1).sort_values(by=['date','group'])

cumulativeData = ordersAggregated.merge(visitsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])

cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visits']

print(cumulativeData.head(5))
```

<div class="alert alert-block alert-success"> <b>Celda [12]</b> <a class="tocSkip"></a> Buena estrategia al calcular métricas acumuladas combinando `orders` y `visits`. Esto permite un análisis claro del rendimiento a lo largo del tiempo.</div> <div class="alert alert-block alert-warning"> <b>Celda [12]</b> <a class="tocSkip"></a> La aplicación de `.apply(lambda x: ...)` puede ser ineficiente en grandes datasets. Considera usar `groupby().agg()` para mejorar el rendimiento.</div>

```python
# 1.- Representa gráficamente el ingreso acumulado por grupo. Haz conclusiones y conjeturas.

cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]

cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]

plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='B')
plt.xticks(rotation=90)
plt.legend()
```

<div class="alert alert-block alert-success"> <b>Celda [13]</b> <a class="tocSkip"></a> Las gráficas de ingresos acumulados son correctas y muestran bien la diferencia entre los grupos.</div> <div class="alert alert-block alert-warning"> <b>Celda [13]</b> <a class="tocSkip"></a> Podrías agregar etiquetas en los ejes y títulos para mejorar la interpretación de las gráficas:</div>

plt.xlabel('Fecha')

plt.title('Ingresos acumulados por grupo')

Podemos ver que el ingreso acumulado de los dos grupos crece de forma gradual pero el del grupo B tuvo un aumento muy importante al rededor del 18 de agosto.

```python
# 2.- Representa gráficamente el tamaño de pedido promedio acumulado por grupo. Haz conclusiones y conjeturas.

cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]

cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]

plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue']/cumulativeRevenueA['orders'], label='A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue']/cumulativeRevenueB['orders'], label='B')
plt.xticks(rotation=90)
plt.legend()
```

En la gráfica podemos ver como el tamaño del pedido promedio del grupo B es considerablemente mayor al del grupo A, a pesar de que empezó con un tamaño menor al pasar de los días lo fue superando, al rededor del día 18 tuvo un aumnento exponencial que luego fue disminuyendo poco a poco.

```python
# Representa gráficamente la diferencia relativa en el tamaño de pedido promedio acumulado para el grupo B en comparación con el grupo A. Haz conclusiones y conjeturas.

mergedCumulativeRevenue = cumulativeRevenueA.merge(cumulativeRevenueB, left_on='date', right_on='date', how='left', suffixes=['A', 'B'])

plt.plot(mergedCumulativeRevenue['date'], (mergedCumulativeRevenue['revenueB']/mergedCumulativeRevenue['ordersB'])/(mergedCumulativeRevenue['revenueA']/mergedCumulativeRevenue['ordersA'])-1)

plt.axhline(y=0, color='black', linestyle='--')
plt.xticks(rotation=90)
```

Podemos ver que los pedidos promedio han sido muy fluctuantes, al principio superó por mucho al grupo A, luego cayó para volver a subir en forma exponencial y al final ha tenido un leve y continuo descenso para situarse aproximadamente un 25% más alto que el grupo A.

```python
# Calcula la tasa de conversión de cada grupo como la relación entre los pedidos y el número de visitas de cada día. Representa gráficamente las tasas de conversión diarias de los dos grupos y describe la diferencia. Saca conclusiones y haz conjeturas.

cumulativeData['conversion'] = cumulativeData['orders']/cumulativeData['visits']

cumulativeDataA = cumulativeData[cumulativeData['group']=='A']

cumulativeDataB = cumulativeData[cumulativeData['group']=='B']

plt.plot(cumulativeDataA['date'], cumulativeDataA['conversion'], label='A')
plt.plot(cumulativeDataB['date'], cumulativeDataB['conversion'], label='B')
plt.legend()
plt.axis([pd.to_datetime("2019-08-01"), pd.to_datetime('2019-09-01'), 0, 0.05])
plt.xticks(rotation=90)
```

```python
# 9.- Encuentra la significancia estadística de la diferencia en la conversión entre los grupos utilizando los datos en bruto. Haz conclusiones y conjeturas.

import scipy.stats as stats

ordersByUsersA = orders[orders['group']=='A'].groupby('visitorId', as_index=False).agg({'transactionId' : pd.Series.nunique})
ordersByUsersA.columns = ['visitorId', 'orders']

ordersByUsersB = orders[orders['group']=='B'].groupby('visitorId', as_index=False).agg({'transactionId' : pd.Series.nunique})
ordersByUsersB.columns = ['visitorId', 'orders']

pd.Series(0, index=np.arange(visits[visits['group']=='A']['visits'].sum() - len(ordersByUsersA['orders'])), name='orders')

sampleA = pd.concat([ordersByUsersA['orders'],pd.Series(0, index=np.arange(visits[visits['group']=='A']['visits'].sum() - len(ordersByUsersA['orders'])), name='orders')],axis=0)

sampleB = pd.concat([ordersByUsersB['orders'],pd.Series(0, index=np.arange(visits[visits['group']=='B']['visits'].sum() - len(ordersByUsersB['orders'])), name='orders')],axis=0)

print("{0:.5f}".format(stats.mannwhitneyu(sampleA, sampleB)[1]), 'El valor p es inferior a 0.05 por lo que rechazamos la hipótesis nula.')

print("{0:.3f}".format(sampleB.mean()/sampleA.mean()-1),'  La ganancia de conversión relativa del grupo B en comparación con el grupo A es del 13.8%')
```

```python
#10 .- Encuentra la significancia estadística de la diferencia en el tamaño promedio de pedido entre los grupos utilizando los datos en bruto. Haz conclusiones y conjeturas.

print('{0:.3f}'.format(stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue'])[1]), 'El valor p es superior a 0.05, por lo que no hay motivo para rechazar la hipótesis nula y concluir que el tamaño promedio de los pedidos difiere entre los grupos.')

print('{0:.3f}'.format(orders[orders['group']=='B']['revenue'].mean()/orders[orders['group']=='A']['revenue'].mean()-1), 'La diferencia relativa es importante, es del 25% del tamaño promedio de pedido entre los grupos')
```

<div class="alert alert-block alert-success"> <b>Celda [21-22]</b> <a class="tocSkip"></a> Bien aplicado el test de Mann-Whitney para comparar conversiones y tamaños de pedido.</div> <div class="alert alert-block alert-warning"> <b>Celda [21]</b> <a class="tocSkip"></a> El formato del `p-value` está bien, pero recuerda que si aplicas múltiples pruebas, deberías **ajustar** los valores `p` usando la corrección de Bonferroni (o cualquier otra de las vistas en la teoría del sprint).</div>

```python
#11.- Encuentra la significancia estadística de la diferencia en la conversión entre los grupos utilizando los datos filtrados. Haz conclusiones y conjeturas.

usersWithManyOrders = pd.concat([ordersByUsersA[ordersByUsersA['orders'] > 2]['visitorId'], ordersByUsersB[ordersByUsersB['orders'] > 2]['visitorId']], axis = 0)
usersWithExpensiveOrders = orders[orders['revenue'] > 435]['visitorId']
abnormalUsers = pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis = 0).drop_duplicates().sort_values()

sampleAFiltered = pd.concat([ordersByUsersA[np.logical_not(ordersByUsersA['visitorId'].isin(abnormalUsers))]['orders'],pd.Series(0, index=np.arange(visits[visits['group']=='A']['visits'].sum() - len(ordersByUsersA['orders'])),name='orders')],axis=0)

sampleBFiltered = pd.concat([ordersByUsersB[np.logical_not(ordersByUsersB['visitorId'].isin(abnormalUsers))]['orders'],pd.Series(0, index=np.arange(visits[visits['group']=='B']['visits'].sum() - len(ordersByUsersB['orders'])),name='orders')],axis=0)

print("{0:.5f}".format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
print("{0:.3f}".format(sampleBFiltered.mean()/sampleAFiltered.mean()-1))
print('El grupo B es mejor que el grupo A, aunque la diferencia no es muy importante.')
```

```python
# 12.- Encuentra la significancia estadística de la diferencia en el tamaño promedio de pedido entre los grupos utilizando los datos filtrados. Haz conclusiones y conjeturas.
print("{0:.3f}".format(stats.mannwhitneyu(
    orders[np.logical_and(
        orders['group']=='A',      np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'],orders[np.logical_and(orders['group']=='B',np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'])[1]))

print("{0:.3f}".format(
    orders[np.logical_and(orders['group']=='B',np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'].mean()  /
    orders[np.logical_and(orders['group']=='A',np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'].mean() - 1))

print('Eliminar los valores atípicos disminuyó la cantidad de ruido en los datos: la varianza disminuyó einclusive ahora vemos que en el grupo B fue 2% menor. El valor p se mantiene mayor al 5%')
```
